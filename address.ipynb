```python
import pandas as pd
import requests
import json

# Define your Google Maps Geocoding API key here
api_key = 'AIzaSyDC1mYfWg0uY2trVbd5m2s6PH9J4n7Lqjo'

def get_lat_lng_from_address(api_key, address):
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    response = requests.get(base_url, params={"address": address, "key": api_key})
    if response.status_code == 200:
        results = response.json().get('results', [])
        if results:
            location = results[0].get('geometry', {}).get('location', {})
            return location.get('lat'), location.get('lng')
    return None, None

def process_addresses(address_list, api_key):
    df = pd.DataFrame(address_list, columns=['Address'])
    df['Latitude'] = None
    df['Longitude'] = None

    for index, row in df.iterrows():
        lat, lng = get_lat_lng_from_address(api_key, row['Address'])
        df.at[index, 'Latitude'] = lat
        df.at[index, 'Longitude'] = lng

    return df

address_list = addresses_list
addresses_df = process_addresses(address_list, api_key)
```
